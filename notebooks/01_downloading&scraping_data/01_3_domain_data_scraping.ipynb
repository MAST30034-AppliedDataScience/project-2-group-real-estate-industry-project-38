```python
from collections import defaultdict
import requests
from bs4 import BeautifulSoup
import re

BASE_URL = "https://www.domain.com.au"
HEADERS = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36'
}
house_url_links = []

property_metadata = defaultdict(dict)

N_PAGES = range(1, 51)
for page in N_PAGES:
    url = BASE_URL + f"/rent/?ptype=house&state=vic&page={page}"
    print(f"In {url}")

    response = requests.get(url, headers=HEADERS)
    
    if response.status_code == 200:
        bs_object = BeautifulSoup(response.content, "lxml")
        print(f"Page {page} parsed successfully.")
        
        index_links = bs_object.find("ul", {"data-testid": "results"}).findAll("a", href=re.compile(f"{BASE_URL}/*"))
        
        for link in index_links:
            if 'address' in link['class']:
                house_url_links.append(link['href'])
    else:
        print(f"Failed to retrieve page {page}: Status code {response.status_code}")
```

```python
import json
import time


# Set request header
HEADERS = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36'
}

property_metadata = defaultdict(dict)

# Get the features we need
for i, property_url in enumerate(house_url_links):
    print(f"Processing {i+1}/{len(house_url_links)}: {property_url}")  

    start_time = time.time()
    try:
        response = requests.get(property_url, headers=HEADERS, timeout=10)
        response.raise_for_status()
        bs_object = BeautifulSoup(response.content, "lxml")
        print(f"Fetched in {time.time() - start_time:.2f} seconds")  

        # Get property name
        property_metadata[property_url]['name'] = bs_object \
            .find("h1", {"class": "css-164r41r"}) \
            .text

        # Get property type
        property_metadata[property_url]['property_type'] = bs_object \
            .find("div", {"data-testid": "listing-summary-property-type"}) \
            .find("span") \
            .text.strip()

        # Get rental price
        property_metadata[property_url]['cost_text'] = bs_object \
            .find("div", {"data-testid": "listing-details__summary-title"}) \
            .text

        # Get coordinates
        coordinates = re.findall(r'destination=([-\d.]+),([-\d.]+)', str(bs_object.find_all("a", {"target": "_blank"})))
        if coordinates:
            latitude, longitude = coordinates[0]
            property_metadata[property_url]['latitude'] = float(latitude)
            property_metadata[property_url]['longitude'] = float(longitude)
        else:
            property_metadata[property_url]['latitude'] = 'N/A'
            property_metadata[property_url]['longitude'] = 'N/A'

        # Get room and parking information
        rooms = bs_object \
            .find("div", {"data-testid": "property-features"}) \
            .findAll("span", {"data-testid": "property-features-text-container"})

        property_metadata[property_url]['rooms'] = [
            re.findall(r'\d+\s[A-Za-z]+', feature.text)[0] for feature in rooms if 'Bed' in feature.text or 'Bath' in feature.text
        ]

        parking_info = [
            re.findall(r'\d+\s[A-Za-z]+', feature.text) for feature in rooms if 'Parking' in feature.text
        ]
        
        if parking_info and parking_info[0]:  
            property_metadata[property_url]['parking'] = parking_info[0][0]
        else:
            property_metadata[property_url]['parking'] = 'N/A'  

        # Get "Date Available" information
        date_available = bs_object.find("ul", {"data-testid": "listing-summary-strip"})
        if date_available:
            available_text = date_available.find("strong")
            if available_text:
                property_metadata[property_url]['date_available'] = available_text.text.strip()
            else:
                property_metadata[property_url]['date_available'] = 'N/A'
        else:
            property_metadata[property_url]['date_available'] = 'N/A'

        # Get describe information
        property_metadata[property_url]['desc'] = re.sub(r'<br/>', '\n', str(bs_object.find("p"))).strip('<p>')

    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {property_url}: {e}")
        continue
    except AttributeError as e:
        print(f"Error parsing data from {property_url}: {e}")
        continue

with open(os.path.join(domain_data_dir, 'house.json'), 'w') as f:
    json.dump(property_metadata, f, indent=4)

print(f"JSON file saved to {os.path.join(domain_data_dir, 'house.json')}")
```
